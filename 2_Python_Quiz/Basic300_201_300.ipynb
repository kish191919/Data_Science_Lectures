```python
def print_arithmetic_operation(a,b):
    print("{} + {} = {}".format(a,b,a+b))
    print("{} - {} = {}".format(a,b,a-b))
    print("{} * {} = {}".format(a,b,a*b))
    print("{} / {} = {}".format(a,b,a/b))
```
